# Multiple Linear Regression

**Problem Statement:**

You are tasked with predicting potato crop yield based on several factors such as temperature, fertilizer usage, and rainfall. Using a dataset of historical data that includes these variables, you will train a multiple linear regression model to predict crop yield. After training, you will evaluate the model's performance by comparing the predicted values with the actual yields from a test dataset.

Your goal is to build a reliable model that helps farmers optimize their practices to maximize yield.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('fertilizers _Multiple Linear.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 1/3)
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

*   List item
*   List item



```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
print(y_pred)
```

```python
plt.scatter(y_test, y_pred)
plt.title('Actual Yield vs Predicted Yield')
plt.plot(y_test, y_test, color = 'red')
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
```
